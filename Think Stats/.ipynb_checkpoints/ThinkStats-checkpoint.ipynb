### Chapter 2: Distributions

One of the best ways to describe a variable is to report the values that appear in the dataset and how many times each value appears. This description is called the distribution of the variable.

The most common representation of a distribution is a histogram
Histogram is a graph that shows the frequency of each value. In this context, 
“frequency” means the number of times the value appears.

```python
# In Python, an efficient way to compute frequencies is with a dictionary. Given a sequence of values, t:
t = [1,3,5,6,7,8,9]
hist = {}
for x in t:
    hist[x] = hist.get(x, 0) + 1
# The result is a dictionary that maps from values to frequencies.
```

```python
# Alternatively, you could use the Counter class defined in the collections module:
from collections import Counter
counter = Counter(t)
# The result is a Counter object, which is a subclass of dictionary.
print(counter)
```

### Representing Histograms

The Hist constructor can take a sequence, dictionary, pandas Series, or another Hist.

```python
hist = thinkstats2.Hist([1,2,2,3,5])
hist
```

```python
type(hist)
```

```python
# Hist objects provide Freq, which takes a value and returns its frequency:
hist.Freq(2)
```

```python
# The bracket operator does the same thing:
hist[2]
```

```python
# If you look up a value that has never appeared, the frequency is 0:
hist[4]
```

```python
# Values returns an unsorted list of the values in the Hist:
hist.Values()
```

```python
# To loop through the values in order, you can use the built-in function sorted:
for val in sorted(hist.Values()):
    print(val, hist.Freq(val))
    
```

```python
# Or you can use Items to iterate through value-frequency pairs:
for val, freq in hist.Items():
    print(val, freq)
```

The expression in brackets is a boolean Series that selects rows from the DataFrame and returns a new DataFrame. Next I generate and plot the histogram of birthwgt_lb for live births.

```python
# Figure 2.1
# Generate and plot the histogram of birthwgt_lb for live births.
hist = thinkstats2.Hist(live.birthwgt_lb, label = 'birthwgt_lb' )
thinkplot.hist(hist)
thinkplot.Show(xlabel ='pounds', ylabel ='frequency');
# The result below. The most common value, called the mode, is 7 pounds. 
# The distribution is approximately bell-shaped, which is the shape of 
#the normal distribution, also called a Gaussian distribution. But unlike 
#a true normal distribution, this distribution is asymmetric; it has a tail
#that extends farther to the left than to the right.
```

When the argument passed to Hist is a pandas Series, any nan values are dropped. label is a string that appears in the legend when the Hist is plotted

```python
# Figure 2.2
hist = thinkstats2.Hist(live.birthwgt_oz, label='birthwgt_oz')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='ounces', ylabel='frequency')
# Figure 2-2 shows the histogram of birthwgt_oz, which is the ounces 
#part of birth weight. 
# In theory we expect this distribution to be uniform; that is, all values 
#should have the same frequency. In fact, 0 is more common than the other values, 
#and 1 and 15 are less common, probably because respondents round off birth weights 
#that are close to an integer value.
```

```python
# Figure 2.3
hist = thinkstats2.Hist(live.agepreg, label='agepreg')
thinkplot.Hist(hist)
thinkplot.Show(xlabel= 'years', ylabel='frequency')
# This figure shows the histogram of agepreg, the mother’s age at the end of pregnancy. 
# The mode is 21 years. The distribution is very roughly bell-shaped, but in this case 
#the tail extends farther to the right than left; most mothers are in their 20s, fewer in their 30s.
```

```python
# Figure 2.4
hist = thinkstats2.Hist(live.prglngth, label='prglngth')
thinkplot.Hist(hist)
thinkplot.Show(xlabel= 'weeks', ylabel='frequency')
# This figure below shows the histogram of prglngth, the length of the pregnancy in weeks. 
# By far the most common value is 39 weeks. The left tail is longer than the right; early 
#babies are common, but pregnancies seldom go past 43 weeks, and doctors often intervene if they do
```
